```python
x1 = [1,2,3]
y1 = [2,4,1]
 
plt.plot(x1, y1, label = "line 1")
 
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
# giving a title to my graph
plt.title('My first line plot')
plt.legend()
plt.show()
```

```python
# Generating x and y points to plot
x1= np.arange(0, 10*(np.pi), 0.1)	
y1 = np.cos(x1)     # cosine function from Numpy library

# potting the points
plt.plot(x1, y1, label = 'cosine') 

plt.legend()              # shows the legend with labels
plt.xlabel('Angle')       # naming the x axis
plt.ylabel('Amplitude')   # naming the y axis

# Generating x and y points for the second plot
x2= np.arange(0, 10*(np.pi), 0.1)   
y2 = np.cos(x2)*0.1*np.arange(10*(np.pi),0,-0.1)         	                     # decaying cosine function

# potting the points
plt.plot(x2, y2, label = 'Decaying cosine')
plt.legend() # show a legend on the plot
plt.title('Two functions on the same graph') # gives a title to the plot

plt.show()	# shows the graph, and removes the text output.
```

```python
# showing multiple bar plots of Iris dataset on the same figure. 

mymap = {'Iris-setosa':0, 'Iris-versicolor':2, 'Iris-virginica':4}
mapped_species = iris_data['Species'].replace(mymap)


plt.bar(mapped_species-0.25/2, iris_data['SepalLengthCm'], color = 'b', width = 0.25, label='SepalLengthCm')
plt.bar(mapped_species+0.25/2, iris_data['SepalWidthCm'], color = 'g', width = 0.25, label='SepalWidthCm')
plt.bar(mapped_species+0.25+0.25/2, iris_data['PetalLengthCm'], color = 'r', width = 0.25, label='PetalLengthCm')
plt.bar(mapped_species+0.5+0.25/2, iris_data['PetalWidthCm'], color = 'k', width = 0.25, label='PetalWidthCm')
plt.legend()
plt.xticks([0, 2, 4], ['setosa', 'versicolor', 'verginica'])
plt.show()
```
